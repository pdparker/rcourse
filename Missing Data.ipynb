Introduction
-------------
<a name="introduction"></a>

R has a huge number of options for missing data. In the R overview of [data cleaning](http://cran.r-project.org/doc/contrib/de_Jonge+van_der_Loo-Introduction_to_data_cleaning_with_R.pdf), 19 distinct packages are avaliable for missing data. Together they cover every concievable approaches to missing values including mean/meadian imputation, LOCF, nearest neighbour, hot deck, random forest, EM, bootstrap EM, and a veriety of multiple imputation procedures.

Today we will focus on the package [Amelia II](http://gking.harvard.edu/amelia) for the following reasons:

1. It is extremely fast
2. I have not known it to crash yet (not converge yes but never crash)
3. Will happily deal with numeric, nominal, or ordinal variables (this is where the speed helps a bunch)
4. Will impute cross-sectional, time series, cross-sectional time series data
5. Really useful diagnostics including the ability to over impute observed variables to see if the missing data model is working

Amealia creates multiple imputations by bootstrapping the EM algorithm. 

Top tips
----------

1. Amealia II is an easy to use imputation package
2. mitools OR Zelig packages provides an easy way of dealing with multiple imputed data
3. semTools or lavaan.survey will provide means of dealing with multiple imputed data in SEM
4. In amelia look at the "chain length" i.e. numbers under --Imputation N -- output. If they are roughly similar for each imputation then the MI has worked well.
5. Working with multiple imputed data requires the use of existing tools AND knowing how to use sapply and lapply
6. Anything you don't want imputed include in the  "idvars" argument in amelia
7. I have not done so here BUT amelia will be considerably faster by running it in parallel (see help file)

```python
%%R
#load required packages
library(Amelia)
library(foreign)
library(mitools)
library(lme4)
#set working directory
setwd("/Users/phparker/Dropbox/Rcourse/data/")
pisa <- read.spss("PISA2003.sav", to.data.frame = TRUE, use.value.labels = FALSE)
#Check percentage of missing data
missing <- apply(pisa, 2, function(x) sum(is.na(x))/length(x)*100)
print(round(missing,3) )
#Here I convert all but ID variables to numeric - You should not have to 
    #Notice the use of NEGATIVE subsetting here. So remove variable i and 2 rather than extract variable 1 and 2
pisa[,-c(1,2)] <- apply(pisa[,-c(1,2)],2, as.numeric)
```

Missing data with Amelia
------------

<a name="amelia"></a>

Here we impute missing data AND over impute one variable to "graphically inspect whether our observed data tends to fall within the region where it would have been imputed had it been missing." [Amelia documentation](http://r.iq.harvard.edu/docs/amelia/amelia.pdf) p.30

```python
%%R
#Missing data model
MI <- amelia(pisa, m = 5, idvars = c(1,2,41,14), noms = c("gender", "IMMIG"), ords = c(3,7:11,15))
# A map of what data was missing
missmap(MI)
#A map of what the imputation looked like
plot(MI, c("SSECATEG", "IMMIG","ESCS" ))
#Lets over impute 
overimpute(MI, var = "ESCS")
```

Analysing Imputed Data
---------------
<a name="analysis"></a>

There are essentially four ways of analysing your muliple imputation data

1. mittools using the with command
2. mittols in combination with the survey package (i.e., What we did for the SEM models)
3. The Zelig package
4. User defined functions

I have used all four in the past. Today we will focus on a combination of 1 and 4.

```python
%%R
#Pass imputations to mitools
pisaMI <- imputationList(MI$imputations)

##Descriptive statistics
#User defined function
MIdescriptives <- function(x){
	temp <- do.call(rbind, x)
	colMeans(temp)
	}

ESCSmean <- with(pisaMI, mean(ESCS))
print(ESCSmean)
cat("The mean of ESCS is\n")
print(MIdescriptives(ESCSmean) )
cat("\n\n")
#Basic Regression
Model1 <- with(pisaMI, lm(SCMAT ~ ESCS*gender))
summary(MIcombine(Model1))
```

Multilevel Models
------------------
<a name="multilevel"></a>

Multilevel models are a little harder in that they require a few additional lines of code. Strucutral equation models we covered by way of exploring plausible values in the previous workshop with lavaan or semtools covering all the heavy lifting.

```python
%%R
#Run the multilevel model
Model2 <- with(pisaMI, lmer(SCMAT ~ ESCS + (1 | SCHOOLID) ) )
#Extract fixed effects
betas<-MIextract(Model2, fun=fixef)
#Extract variance covaraince matrix
    #Note if using nlme rather than lme4 package you only need MIextract(Model2, fun=vcov)
vars<- lapply(MIextract(Model2, fun=vcov), as.matrix)
#Print the output
summary(MIcombine(betas,vars))
```
