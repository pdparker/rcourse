Top Tips for running regression
---------------------------------
<a name="regression"></a>

1. Almost ALL model functions will have the following structure $response \sim predictor + predictor$
2. Once models are run you can pass them to summary() OR plot() OR residuals() OR coef() OR confint() OR predict() OR lm.beta()
3. Standardized betas come from the lm.beta() function that is in the QuantPsyc package
4. Interactions can be defined in models short hand via $response \sim predictor * predictor$ or long hand $response \sim predictor + predictor + predictor:predictor$ 
5. You can make ANY transformation you want directly in the model including polynomials, interactions, log transformations, z-scores
6. To compare the fit of nested models use anova()
7. To calculate simple slopes use deltaMethod() in the car package
8. To calculate mediation use the mediation package

Regression is run with the command lm. Associated with this is glm for logistic regression, t.test, and aov for anova's.

All this models have a straightforward syntax for building a model:

responseVariable ~ predictor1 + predictor2, data=theDataYouAreUsing, additionalOptions.

Lets start by looking at the effects of different achievement domains on self-concept.

```python
%%R
#Run ther model and save it to an object
model1 <- lm(SCMAT ~ PV1MATH + PV1READ + PV1SCIE, data = pisa)
#Return the output
print(model1)
print(summary(model1))
#Check the assumptions
plot(model1)
```

Notice how small the coefficents are. This is becuase achievement is on a very large scale. So we should fix this somehow. We could:

* Z score the achievement scores
* Devide them by a large number

We also have to choices on how to do this:

* On the fly in the model command
* Change the variables

[top](#top)

```python
%%R
# Here are our original coefficents
cat("Original coeficents:\n")
print(coef(model1))
cat("\n\n")
#Z score on the fly
model2 <- lm(SCMAT ~ scale(PV1MATH) + scale(PV1READ) +
             scale(PV1SCIE), data = pisa)
#Return the output
cat("Z-scored coeficents:\n")
print(coef(model2))
cat("\n\n")
# Change metric on the fly
model3 <- lm(SCMAT ~ I(PV1MATH/100) + I(PV1READ/100) +
             I(PV1SCIE/100), data = pisa)
#Return the output
cat("Scaled coeficents:\n")
print(coef(model3))
cat("\n\n")
# we could also get the standardized beta's
library(QuantPsyc)
cat("Standardized coeficents:\n")
print(lm.beta(model1))
```

Polynomials

<a name="poly"></a>

Maybe achievement predicts SCMAT differently at either end of the achievement spectrum. So lets try a polynomial model.

***Note***: In the poly function below raw = TRUE is equivilent to transforming data by hand (e.g. I(SCMAT^2) ). raw = FALSE will give the orthoginal polynomial contrasts

[top](#top)

```python
%%R
#Second order polynomial
modelQuad <- lm(SCMAT ~ PV1MATH + I(PV1MATH^2) , data = pisa)
print(summary(modelQuad) )
#Third Order polynomial
modelPoly <- lm(SCMAT ~ poly(PV1MATH, degree = 3, raw = TRUE) , data = pisa)
print(summary(modelPoly) )
print(coef(modelPoly) )
temp <- coef(modelQuad)
ployFun <- function(x) {temp[1] + temp[2]*x + temp[3]*x^2 }
curve(ployFun,-2,2, col = "blue", xlim = c(-2,3) )
text(2.5,1, "Quadratic\nEffect", col = "blue")
```

Mediation
-------------
<a name="mediation"></a>

Mediation is used a lot in the sort of research we do. Here we are going to look at a general purpose approach to mediation that will work regardless of whether your models are regression, multilevel, logistic, or polr (this is a critical strength!). To do this we will use the mediation package. The good thing about mediation is that it will produce estimates via boot strapping or simulation thus overcoming some complaints about the Barron and Kenny method.

The question we will ask is the degree to which the effect of gender on SCMAT is mediated by achievement.

[top](#top)

```python
%%R
#load the mediation package
library(mediation)
# here we will use the subset command to account for missing data on SCMAT
# First we explore the effect of gender on achievement
modelA <- lm(PV1MATH ~ gender, data = pisa, subset = !is.na(SCMAT) )
# Second we estimate the effect of gender and PV1MATH allowing for a treatment by mediator interaction
modelB <- lm(SCMAT ~ gender*PV1MATH, data = pisa, subset = !is.na(SCMAT) )
# Now we calculate the mediation using simulation (see help file for how to do bootstrapping)
modelAB <- mediate(modelA, modelB, treat = 'gender', mediator = 'PV1MATH', sims = 100)
# Get the results and plot them
print(summary(modelAB) )
plot(modelAB)
```

General Linear models

<a name="glm"></a>


Top Tips for GLM
-------------------

1. For binary variables use family=binomial(link = "logit") for logistic regression or family=binomial(link = "probit") for probit regression
2. For counts use family = "poisson"
3. Many other data types can be covered. See ?family for details


Running general linear models
------------------------------
So far we have assumed that our response variable is continous. But this is not always the case. When response variables are ordinal, nominal, binomial, or count we need to turn to general linear models. 

In this section we are going to look at how strong the relationship is between objective measures of social class and self-report measures. We will run a logistic regression model and then a proportional odds regression model.

[top](#top)

```python
%%R
#First lets create a blue colar versus white colar variable
# We will do this via the recode function in the car package
library(car)
pisa$socialClass <- recode(pisa$SSECATEG, "c(1,2) = 1; c(3,4) = 0")
# now we can run our logistic regression
modelLogistic <- glm(socialClass ~ ESCS, data = pisa, family = binomial(link = 'logit') )
# Summary of the model
print(summary(modelLogistic) )
# Coefs in log odds
cat("\nLog odds are:\n\n")
print(exp(coef(modelLogistic) ) )
```

***Note:*** It is also easy to run a probit instead of logistic. Just change link='logit' to link='probit'

Proportional Odds Logistic Regression
-------------------------------------
<a name="polr"></a>

The above model is fine BUT we are reducing the amount of detail in the data as there are four categories of self-reported social class rather than 2. To account for this ordinal data we can use the polr  function in the MASS package. 

[top](#top)

```python
%%R
library(MASS)
# we need to r
modelPolr <- polr(as.factor(rev(SSECATEG)) ~ ESCS, data = pisa)
print(summary(modelPolr))

cat("Log odds are:\n")
print(exp(modelPolr$zeta))
print(exp(coef(modelPolr)))
```

Multilevel Models
===================
<a name="multilevel"></a>

Top tips for multilevel models
-------------------------------

1. ICC1 and ICC2 functions in the multilevel package are easy ways to get ICCs
2. Multilevel models look exactly like lm or glm models except for the addition of a random component
3. The random component is an additional predictor and looks like (bitYouWantRandom | cluster)
4. 1 = intercept so a random intercept for school id would look like (1 | schoolId)
5. This structure is identical for either lm or glm models
6. Three or more level nesting looks like (1 | classID/schoolID)
7. Cross classification looks like (1 | schoolID) + (1 | neighbourhoodID)
8. You can use a hausman endogineity test to see whether you even need a multilevel model
9. You can compare nested multilevel models using anova()
10. Multilevel models are run from the lme4 package, however if you need to account for autoregressive errors you will need to use the nlme package

Multilevel Models
---------------
We will look at whether we need a multilevel model to explore the relationship between achievement and self-concept using a HET. Then we compare a random intercept to a random slopes model. There is no real function for a HET in R so we will have to slightly modify an existing one taken from [stackoverflow](https://stackoverflow.com/questions/23630214/hausmans-specification-test-for-glmer-from-lme4).

[top](#top)

```python
%%R
#First we need to write a function for the HET
phtest_glmer <- function (glmerMod, glmMod, ...)  {  ## changed function call
  coef.wi <- coef(glmMod)
  coef.re <- fixef(glmerMod)  ## changed coef() to fixef() for glmer
  vcov.wi <- vcov(glmMod)
  vcov.re <- vcov(glmerMod)
  names.wi <- names(coef.wi)
  names.re <- names(coef.re)
  coef.h <- names.re[names.re %in% names.wi]
  dbeta <- coef.wi[coef.h] - coef.re[coef.h]
  df <- length(dbeta)
  dvcov <- vcov.re[coef.h, coef.h] - vcov.wi[coef.h, coef.h]
  stat <- abs(t(dbeta) %*% as.matrix(solve(dvcov)) %*% dbeta)  ## added as.matrix()
  pval <- pchisq(stat, df = df, lower.tail = FALSE)
  names(stat) <- "chisq"
  parameter <- df
  names(parameter) <- "df"
  alternative <- "one model is inconsistent"
  res <- list(statistic = stat, p.value = pval, parameter = parameter, 
              method = "Hausman Test",  alternative = alternative,
              data.name=deparse(glmerMod@call$data))  ## changed
  class(res) <- "htest"
  return(res)
}
#Load the lme4 package
library(lme4)
#Fixed effect model
modelFixed <- lm(SCMAT ~ PV1MATH, data = pisa)
#Random intercept model
modelRandomInt <- lmer(SCMAT ~ PV1MATH + (1|SCHOOLID.x) , data = pisa)
# Test whether a multilevel model is needed
print(phtest_glmer(modelRandomInt, modelFixed) )
# Ok so we don't really need it in this case we could get away with just using complex standard errors 

# Lets fit a random slope model
modelRandomSlope <- lmer(SCMAT ~ PV1MATH + (1|SCHOOLID.x) + (PV1MATH|SCHOOLID.x) , data = pisa)

#Lets compare a random intercept to a random slope model
print(anova(modelRandomInt, modelRandomSlope))

# Print output of the Random intercept model
print(summary(modelRandomSlope) )
```

I have modified a function  from [here](http://www.rensenieuwenhuis.nl/r-sessions-20-plotting-multilevel-models/) that will give a nice plot of the multilevel model.

```python
%%R
#plot
visualize.lme <- function (model, coefficient, group, col.line = "black", ...){
        r <- ranef(model)[[group]]
        f <- fixef(model)

        effects <- data.frame(r[,1]+f[1], r[,2]+f[2])

        number.lines <- nrow(effects)

        predictor.min <- tapply(model@frame[[coefficient]], model@frame[[group]], min, na.rm=TRUE)
        predictor.max <- tapply(model@frame[[coefficient]], model@frame[[group]], max, na.rm=TRUE)

        outcome.min <- min(predict(model), na.rm=TRUE)
        outcome.max <- max(predict(model), na.rm=TRUE)

        plot (c(min(predictor.min),max(predictor.max)),c(outcome.min,outcome.max), type="n", ...)

        for (i in 1:number.lines){
            expression <- function(x) {effects[i,1] + (effects[i,2] * x) }
            curve(expression, from=predictor.min[i], to=predictor.max[i], add=TRUE, col = col.line)
            }
        }

visualize.lme(modelRandomSlope, "PV1MATH", "SCHOOLID.x", 
              xlab="Achievement in Math", ylab="self-concept",
              main="Effect of achievement on self-concept stratified by school", col.line = "#30303080")
```
