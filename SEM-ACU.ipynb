Set up of survey data
----------------------
<a name="complex"></a>

Data like PISA uses a complex sampeling procedure. In the case of PISA this procedure has two stages:

1. Selection of school proportional to size
2. Selection of participants from within that school

To account for this we sampeling we may want to use multilevel modelling but for SEM in our case we will just focus on fixed effects. The PISA manual thus suggests we should analyze the data by a) using population weights, b) using replicant weights to account for the complex data structure and c) plausible values for achievement.

To account for all this in a model is complex so we are going to off-load the heavy lifting to the lavaan survey package.


```python
%%R
#set working directory
setwd("/Users/phparker/Dropbox/Rcourse/data")
#read in the data
library(foreign)
pisa <- read.spss("PISA2003-items.sav", to.data.frame=TRUE, use.value.labels = FALSE)
#load required packages
library(lavaan)
library(semTools)
###Set up survey data
```

How to Run SEM
---------------
To run a CFA or SEM model we will use the lavaan package. To account for survey design we will use lavaan survey. Our general approach will be:

1. Write a model
2. Pass it to lavaan
3. Extract the information you want

The model syntax:

1. Factor loadings:  =~
2. Regression, intercepts, and latent means:  ~
3. Covariances/variances we use ~~
4. Naming OR constraining parameters: name*parameter 
5. Fixing a parameter: number*parameter
6. Start values: start(number)*parameter
7. Delta Method (e.g, indirect effects): newParameter := parameter*parameter

Lavaan magic
--------------

Fitting a CFA
--------------
<a name="cfa"></a>

Here we fit a CFA model:

```python
%%R
#Model syntax
model <- '
        selfConcept =~ ST32Q02 + ST32Q04 + ST32Q06 + ST32Q07 + ST32Q09
        anxiety =~ ST32Q01 + ST32Q03 + ST32Q05 + ST32Q08 + ST32Q10
        '
# Model without population corrections
fit <- cfa(model, data=pisa, estimator = 'MLR')
#Compare fit
cat("fit indices for uncorrected data:\n")
print(fitMeasures(fit)[c("tli", "cfi", "rmsea", "chisq", "df")])
# Standardized estimates
#or summary to get everything
cat("Standardized parameter estimates:\n")
print(standardizedSolution(fit) )
```

Measurement Invariance
------------------------
<a name="invariance"></a>

Measurement Invariance aims to explore whether a measure or set of relationships are equivilent across groups. For example, is our measure of self-concept the same for boys and girls. This allows us to distinguish between differences in 'kind' (self-concept means different things) and 'degree' (difference levels of endorsement on the same type of self-concept).

To run invariance you can add group="groupingVariables";
And then you fix parameters you add the key words group.equal=c("loadings","intercepts")
to the original cfa() function.

Or you can use the semTools package to automate either a four (strict=FALSE) or five (strict=TRUE) model invariance.

```python
%%R
## Multiple groups
measurementInvariance(model, data=pisa, estimator = 'MLR', missing = 'default', group = 'gender')
## NOTE: THe new version of measurement invariance seems to be buggy. Check progression of df values
```

Structural Equation Modelling
--------------------------------
<a name="sem"></a>

Here is an example mplus model with output set to mimic Mplus

```python
%%R
## Structural equation modelling
model <-  '
selfConcept =~ ST32Q02 + ST32Q04 + ST32Q06 + ST32Q07 + ST32Q09
anxiety =~ ST32Q01 + ST32Q03 + ST32Q05 + ST32Q08 + ST32Q10
selfConcept ~ PV1MATH + PV1READ + PV1SCIE
anxiety ~ PV1MATH + PV1READ + PV1SCIE
'
# Model without population corrections
fit <- sem(model, data=pisa, estimator = 'MLR', missing = 'default')
# Summary statistics
summary(fit, fit.measures = TRUE, standardized = TRUE)
```

SEM - Multigroup
-----------------
<a name="semInvariance"></a>

Here is an example multigroup SEM. Here we use the power of R's regular expresions to extract ONLY the regression estimates for boys and girls.

```python
%%R
## Multigroup SEM
fitFree <- sem(model, data=pisa, estimator = 'MLR', missing = 'default', group = 'gender',
		   group.equal = c("loadings"))
fitFixed <- sem(model, data=pisa, estimator = 'MLR', missing = 'default', group = 'gender',
		   group.equal = c("loadings","regressions"))
print(anova(fitFree, fitFixed))
# Summary statistics
out <- standardizedsolution(fitFree)
regression <- intersect(grep("(MATH|READ|SCIE)", out$rhs), grep("(selfConcept|anxiety)", out$lhs) )
print(out[regression,])
```

Additional functions
---------------------
<a name="additional"></a>

semTools together with lavaan will do pretty much anything including power analysis, power analysis for multiple groups, compare fitted models, deal with multiple imputations, longInvariance (for longitudinal invariance), latent interactions and ploting of those interactions, imposeStart (same as svalues in mplus), net (Test whether models are equivalent or nested)

The [simsem](http://simsem.org/) package will run SEM simulations via lavaan

[xxM](http://xxm.times.uh.edu/) will run multilevel SEM. It uses either LISREL style syntax or AMOS style graphics based model development.

[openMX](http://openmx.psyc.virginia.edu/) will run everything (mixture, latent profile, growth mixture, multilevel, everything else).
